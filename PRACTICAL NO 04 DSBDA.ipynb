```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Convert to DataFrame
xtrain = pd.DataFrame(xtrain)
xtest = pd.DataFrame(xtest)

# Apply get_dummies
xtrain = pd.get_dummies(xtrain)
xtest = pd.get_dummies(xtest)

# Save columns
xtrain_columns = xtrain.columns

# Align xtest to xtrain
xtest = xtest.reindex(columns=xtrain_columns, fill_value=0)

# Replace inf
xtrain = xtrain.replace([np.inf, -np.inf], np.nan)
xtest = xtest.replace([np.inf, -np.inf], np.nan)

# Impute
imputer = SimpleImputer(strategy="mean")
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)

# Train
lm = LinearRegression()
lm.fit(xtrain, ytrain)

# Predict
ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)

print("All good ✅")
```

```python
from sklearn.metrics import r2_score

# Predictions
ytrain_pred = lm.predict(xtrain)
ytest_pred = lm.predict(xtest)

# Fix shape if needed
ytest = ytest.ravel()

# Scores
print("Train R2:", r2_score(ytrain, ytrain_pred))
print("Test R2:", r2_score(ytest, ytest_pred))
```
