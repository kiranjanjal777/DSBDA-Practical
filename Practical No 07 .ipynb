```python
processed_docs = []

for doc in documents:
    doc = doc.lower()
    words = word_tokenize(doc)
    filtered_words = [w for w in words if w not in stop_words]
    processed_docs.append(filtered_words)

print("Processed Documents:")

for i, doc in enumerate(processed_docs):
    print(f"Doc {i+1}: {doc}")
```

```python
vocab = set()
for doc in processed_docs:
    vocab.update(doc)
vocab = list(vocab)
print("\nVocabulary:", vocab)
```

```python
def compute_tf(doc):
    tf_dict ={}
    doc_len = len(doc)
    print("\nTF Calculation:")
    for word in vocab:
        count = doc.count(word)
        tf = count / doc_len
        tf_dict[word]= tf
        print(f"{word} = {count}/{doc_len} = {tf}")
        return tf_dict
tf_docs =[]
for i, doc in enumerate(processed_docs):
    print(f"\n--- TF for Document {i+1} ---")
    tf_docs.append(compute_tf(doc))
```

```python
from collections import Counter

# Step 1: Define document
docA = "Tokenization is the first step in text analytics"

# Step 2: Split into words
bagOfWordsA = docA.split()

# Step 3: Count frequency
wordDictA = Counter(bagOfWordsA)

# Step 4: Count total words
numOfWordsA = len(bagOfWordsA)

# Output
print("Bag of Words:", bagOfWordsA)
print("Word Frequency:", wordDictA)
print("Total Words:", numOfWordsA)
```

```python
def computeTF(wordDict, totalWords):
    tfDict = {}

    print("Word Counts:", wordDict)
    print("\nTF Values:\n")

    for word, count in wordDict.items():
        tf = count / float(totalWords)
        tfDict[word] = tf

    return tfDict
```

```python
tfA = computeTF(wordDictA, numOfWordsA)
```

```python
import math

def compute_idf(docs):
    N = len(docs)
    idf_dict = {}

    print("\nIDF Calculation:")

    # Create vocabulary (unique words)
    vocab = set(word for doc in docs for word in doc)

    for word in vocab:
        count = sum(1 for doc in docs if word in doc)
        idf = math.log(N / (count + 1)) + 1   # smoothing
        idf_dict[word] = idf

        print(f"{word} → appears in {count} docs → IDF = {idf}")

    return idf_dict


# Call function
idf = compute_idf(processed_docs)
```

```python
def compute_tfidf(tf, idf):
    tfidf = {}

    for word in tf:   # use words from tf instead of global vocab
        tfidf[word] = tf[word] * idf[word]

    return tfidf


tfidf_docs = []

for i, tf in enumerate(tf_docs):
    tfidf_docs.append(compute_tfidf(tf, idf))

print("\nTF-IDF Results:")

for i, tfidf in enumerate(tfidf_docs):
    print(f"\nDocument {i+1}:")
    
    for word, value in tfidf.items():
        print(f"{word}: {value}")
```
