```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Solar ":[1200,1500,1300],
    "Wind":[3400,3600,3200],
    'Hydropower':[2900,3100,2800],
    "Biomas":[2600,2700,2400]
}
df=pd.DataFrame(data)
```

```python
# Create a pair plot
sns.pairplot(df)

# Show the plot
plt.show()
```

```python
# Create a heatmap to show the Coreeltion
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidths=0.5)

#Add title
plt.title('Correlation Heatmap between energy Source')
# Show the plot
plt.show()
```

```python
#Create a Box plot
sns.boxplot(data=df)

#Add title
plt.title('Box Plot of Energy Sources')
plt.show()
```

```python
#VolinPlot
sns.violinplot(data=df)

#Add title
plt.title('Violin Plot of Energy Sources')
plt.show()
```

```python
#sample data for energy consumption and carbon emission
energy_consumption = [100, 200 ,300,400,500]
carbon_emission = [10, 20,30,40,50]

#Create a datafram
df_reg=pd.DataFrame({'Energy Consumption':energy_consumption,'Carbon Emission':carbon_emission})
#create a regression plot
sns.regplot(data=df_reg,x='Energy Consumption',y='Carbon Emission')
#Add Title
plt.title('Regression Plot of Energy Consumption vs Carbon Emission')
#Show the plot
plt.show()
```

```python
# Define the months
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Create dataset
df_facet = pd.DataFrame({
    'Month': months * 3,
    'Energy Consumption': [
        1200,1300,1100,1500,1400,1600,1250,1350,1150,1550,1450,1650,
        1100,1200,1000,1400,1300,1500,1150,1250,1050,1450,1350,1550,
        1300,1400,1200,1600,1500,1700,1350,1450,1250,1650,1550,1750
    ],
    'Region': ['North']*12 + ['South']*12 + ['East']*12
})

# Create FacetGrid plot
g = sns.FacetGrid(df_facet, col='Region', height=4, aspect=1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')

plt.show()
```
