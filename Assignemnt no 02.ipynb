```python
series = pd.isnull(df["math_score"])
df[series]
```

```python
df['math_score'] = df['math_score'].fillna(df['math_score'].mean())
```

```python
df.isnull().sum()
```

```python
df['reading_score'] = pd.to_numeric(df['reading_score'], errors='coerce')
```

```python
df['reading_score'] = df['reading_score'].fillna(df['reading_score'].mean())
```

```python
import matplotlib.pyplot as plt

df.boxplot()

plt.show()
```

```python
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(df[['math_score']]))

print(z)
```

```python
threshold = 3
```

```python
print(np.where(z > threshold))
```

```python
df_new = df[(z < threshold).all(axis=1)]

df_new
```

```python
df_new.boxplot()

plt.show()
```

```python
plt.scatter(df_new['performance_score'], df_new['math_score'])

plt.xlabel("Performance Score")

plt.ylabel("Math Score")

plt.show()
```

```python
Q1 = df['math_score'].quantile(0.25)

Q3 = df['math_score'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

print(lower, upper)
```

```python
outliers = df[(df['math_score'] < lower) | (df['math_score'] > upper)]

outliers
```

```python
df_clean = df[(df['math_score'] >= lower) & (df['math_score'] <= upper)]

df_clean
```

```python
df['Duration'] = 2024 - df['club_joindate']

df
```
